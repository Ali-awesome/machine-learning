### Importing packages and loading data.

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('./data.csv')
df.head()
```

#### Making the data clean and consistant

```python
# As all the columns are not consistant in their appearance so we will make them consistant looking.
# Making the column heading consitant looking
df.columns = df.columns.str.lower().str.replace(' ','_')
df.head(2)
```

```python
# Making the values consistant
listOfStrCol = list(df.dtypes[df.dtypes == 'object'].index)
# df.dtypes[df.dtypes == 'object'].index
listOfStrCol
for col in listOfStrCol:
    df[col] = df[col].str.lower().str.replace(' ','_')
df.head()
# df[listOfStrCol[0]].str.lower().str.replace(" ","_")
```

```python
null_val = df.isna().sum()
null_val.index
null_val_df = pd.DataFrame(null_val, index=null_val.index, columns = ['Values'])
bar = null_val_df[null_val_df.Values != 0].plot.bar()
bar.bar_label(bar.containers[0])
# null_val_df.columns = ['Missing_Values']
# NV_df = null_val_df[null_val_df.Missing_Values != 0]
# # for val in NV_df.columns[0]:
# NV_df.plot.bar()
```
